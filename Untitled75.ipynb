```python

# ABET Compliance Checking Agent - Customized for Your PDF
# This notebook is modified to work with your specific PDF file: "ABET Self Study Report ME final.pdf"

# Install required packages
!pip install -q transformers pdfplumber spacy matplotlib

# Install spaCy language model
!python -m spacy download en_core_web_sm

# Import required libraries
import pandas as pd
import numpy as np
import re
import json
import os
import torch
from datetime import datetime
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from sklearn.model_selection import train_test_split
import spacy
import matplotlib.pyplot as plt
import pdfplumber
from google.colab import files
import warnings
warnings.filterwarnings('ignore')
```

```python
class ABETComplianceAgent:
    """Main class for Agent 2 - Compliance Checking Agent"""

    def __init__(self):
        """Initialize the agent with necessary components"""
        self.nlp = spacy.load("en_core_web_sm")
        self.criteria_labels = {
            'A': 'Student Outcomes',
            'B': 'Continuous Improvement',
            'C': 'Curriculum',
            'D': 'Faculty',
            'E': 'Facilities',
            'F': 'Institutional Support',
            'G': 'Program Criteria',
            'H': 'Student Support'
        }

        # Rules engine for ABET criteria compliance
        self.rules = self._load_default_rules()

        # Output structures
        self.classified_paragraphs = None
        self.compliance_scores = None

    def _load_default_rules(self):
        """Load default ABET rules for evaluation"""
        rules = {
            'A': {
                'required_keywords': ['student outcomes', 'learning outcomes', 'competencies', 'skills', 'abilities'],
                'min_evidence': 5,
                'importance': 0.9
            },
            'B': {
                'required_keywords': ['continuous improvement', 'assessment', 'evaluation', 'feedback', 'measure'],
                'min_evidence': 4,
                'importance': 0.8
            },
            'C': {
                'required_keywords': ['curriculum', 'courses', 'syllabus', 'content', 'requirements'],
                'min_evidence': 6,
                'importance': 0.9
            },
            'D': {
                'required_keywords': ['faculty', 'professors', 'instructors', 'qualification', 'expertise'],
                'min_evidence': 3,
                'importance': 0.7
            },
            'E': {
                'required_keywords': ['facilities', 'laboratories', 'equipment', 'resources', 'infrastructure'],
                'min_evidence': 4,
                'importance': 0.6
            },
            'F': {
                'required_keywords': ['institutional', 'support', 'budget', 'resources', 'administration'],
                'min_evidence': 3,
                'importance': 0.5
            },
            'G': {
                'required_keywords': ['program criteria', 'specific requirements', 'domain-specific', 'discipline'],
                'min_evidence': 4,
                'importance': 0.8
            },
            'H': {
                'required_keywords': ['student support', 'advising', 'counseling', 'services', 'resources'],
                'min_evidence': 3,
                'importance': 0.6
            }
        }
        return rules
```

```python


    def extract_paragraphs_from_pdf(self, pdf_path):
        """Extract paragraphs from PDF files"""
        paragraphs = []
        try:
            with pdfplumber.open(pdf_path) as pdf:
                for page_num, page in enumerate(pdf.pages):
                    text = page.extract_text() or ""
                    # Clean up text
                    text = text.replace('\n\n', '||PARAGRAPH_BREAK||')
                    text = text.replace('\n', ' ')
                    text = text.replace('||PARAGRAPH_BREAK||', '\n\n')

                    # Split text into paragraphs
                    page_paragraphs = re.split(r'\n\n+', text)

                    # Filter out too short paragraphs and add to collection
                    filtered_paragraphs = []
                    for p in page_paragraphs:
                        p = p.strip()
                        if len(p.split()) > 15:  # Only paragraphs with more than 15 words
                            filtered_paragraphs.append({
                                'text': p,
                                'page': page_num + 1
                            })

                    paragraphs.extend(filtered_paragraphs)

            print(f"Extracted {len(paragraphs)} paragraphs from {pdf_path}")
            return paragraphs
        except Exception as e:
            print(f"Error extracting paragraphs from PDF: {str(e)}")
            return []

    def rule_based_classification(self, paragraphs):
        """Classify paragraphs based on keyword rules when no model is available"""
        results = []

        for para_info in paragraphs:
            paragraph = para_info['text']
            page = para_info['page']

            # Initialize scores for each criterion
            scores = {}
            paragraph_lower = paragraph.lower()

            # Calculate score for each criterion based on keyword presence
            for criterion, rule in self.rules.items():
                score = 0
                keywords = rule['required_keywords']

                # Count keyword matches
                keyword_matches = 0
                for keyword in keywords:
                    if keyword.lower() in paragraph_lower:
                        keyword_matches += 1

                # Calculate score based on keyword matches (normalized by number of keywords)
                score = keyword_matches / len(keywords) if keywords else 0
                scores[criterion] = score

            # Assign to criterion with highest score if it's above threshold
            max_criterion = max(scores, key=scores.get)
            confidence = scores[max_criterion]

            # Only assign if confidence is above threshold
            if confidence >= 0.2:  # Adjust threshold as needed
                assigned_criterion = max_criterion
                criterion_name = self.criteria_labels[assigned_criterion]
            else:
                # If no clear match, mark as unknown
                assigned_criterion = 'Unknown'
                criterion_name = 'Unknown'
                confidence = 0.0

            results.append({
                'paragraph': paragraph,
                'page': page,
                'criterion': assigned_criterion,
                'criterion_name': criterion_name,
                'confidence': confidence
            })

        self.classified_paragraphs = pd.DataFrame(results)
        return self.classified_paragraphs

    def evaluate_compliance(self, min_confidence=0.2):
        """Evaluate compliance with ABET criteria based on classified paragraphs"""
        if self.classified_paragraphs is None:
            raise ValueError("No classified paragraphs available. Run classify_paragraphs first.")

        # Filter paragraphs with confidence above threshold
        df = self.classified_paragraphs[self.classified_paragraphs['confidence'] >= min_confidence]
        df = df[df['criterion'] != 'Unknown']  # Exclude unknown criteria

        # Calculate compliance scores for each criterion
        scores = {}
        for criterion, name in self.criteria_labels.items():
            criterion_df = df[df['criterion'] == criterion]

            # Count evidence items for this criterion
            evidence_count = len(criterion_df)
            required_count = self.rules[criterion]['min_evidence']

            # Calculate coverage score
            coverage = min(evidence_count / required_count, 1.0) if required_count > 0 else 0

            # Calculate quality score based on confidence values
            quality = criterion_df['confidence'].mean() if not criterion_df.empty else 0

            # Apply rule-based evaluation
            rule_score = self._apply_rules(criterion, criterion_df)

            # Calculate final score (weighted average)
            final_score = 0.7 * coverage + 0.2 * quality + 0.1 * rule_score

            scores[criterion] = {
                'name': name,
                'coverage': coverage,
                'quality': quality,
                'rule_score': rule_score,
                'score': final_score,
                'evidence_count': evidence_count,
                'required_count': required_count,
                'status': 'Pass' if final_score >= 0.7 else 'Fail'
            }

        # Calculate overall score
        overall_score = sum(item['score'] * self.rules[crit]['importance']
                            for crit, item in scores.items()) / sum(rule['importance'] for rule in self.rules.values())

        # Create final compliance report
        self.compliance_scores = {
            'criteria': scores,
            'overall': {
                'score': overall_score,
                'status': 'Ready' if overall_score >= 0.75 else 'Needs Improvement' if overall_score >= 0.6 else 'Defer',
                'last_update': datetime.now().strftime("%Y-%m-%dT%H:%M:%SZ")
            }
        }

        return self.compliance_scores

    def _apply_rules(self, criterion, paragraphs_df):
        """Apply specific rules for criterion evaluation"""
        if paragraphs_df.empty:
            return 0

        # Get rule keywords for this criterion
        keywords = self.rules[criterion]['required_keywords']

        # Check for keyword presence in paragraphs
        keyword_matches = 0
        for paragraph in paragraphs_df['paragraph']:
            paragraph_lower = paragraph.lower()
            for keyword in keywords:
                if keyword.lower() in paragraph_lower:
                    keyword_matches += 1
                    break

        # Calculate rule score based on keyword matches
        rule_score = keyword_matches / len(paragraphs_df) if len(paragraphs_df) > 0 else 0

        return rule_score

    def generate_compliance_report(self, output_path=None):
        """Generate a compliance report with scores and non-compliance items"""
        if self.compliance_scores is None:
            raise ValueError("No compliance scores available. Run evaluate_compliance first.")

        # Create criteria scores JSON
        report = {
            'overall': self.compliance_scores['overall'],
            'criteria': {}
        }

        for criterion, data in self.compliance_scores['criteria'].items():
            report['criteria'][criterion] = {
                'name': data['name'],
                'coverage': round(data['coverage'], 2),
                'quality': round(data['quality'], 2),
                'score': round(data['score'], 2),
                'status': data['status']
            }

        # Create non-compliance report (for criteria that failed)
        non_compliance = []
        for criterion, data in self.compliance_scores['criteria'].items():
            if data['status'] == 'Fail':
                non_compliance.append({
                    'criterion': criterion,
                    'name': data['name'],
                    'gap_type': 'Insufficient Evidence' if data['coverage'] < 0.7 else 'Poor Quality',
                    'evidence_needed': max(0, data['required_count'] - data['evidence_count']),
                    'priority': self.rules[criterion]['importance']
                })

        # Sort by priority
        non_compliance = sorted(non_compliance, key=lambda x: x['priority'], reverse=True)

        # Save reports if output path is provided
        if output_path:
            os.makedirs(output_path, exist_ok=True)

            # Save criteria scores
            with open(os.path.join(output_path, 'criteria_scores.json'), 'w') as f:
                json.dump(report, f, indent=2)

            # Save non-compliance report
            non_compliance_df = pd.DataFrame(non_compliance)
            if not non_compliance_df.empty:
                non_compliance_df.to_csv(os.path.join(output_path, 'non_compliance_report.csv'), index=False)

            # Save classified paragraphs with their criteria
            if self.classified_paragraphs is not None:
                self.classified_paragraphs.to_csv(os.path.join(output_path, 'classified_paragraphs.csv'), index=False)

            print(f"Reports saved to {output_path}")

        return report, non_compliance

    def visualize_results(self, output_path=None):
        """Visualize compliance results"""
        if self.compliance_scores is None:
            raise ValueError("No compliance scores available. Run evaluate_compliance first.")

        # Create dataframe for plotting
        plot_data = []
        for criterion, data in self.compliance_scores['criteria'].items():
            plot_data.append({
                'criterion': f"{criterion} - {data['name']}",
                'coverage': data['coverage'],
                'quality': data['quality'],
                'score': data['score']
            })

        df = pd.DataFrame(plot_data)

        # Create score visualization
        plt.figure(figsize=(14, 8))

        # Plot bars for each metric
        bar_width = 0.25
        x = np.arange(len(df))

        plt.bar(x - bar_width, df['coverage'], width=bar_width, label='Coverage', color='skyblue')
        plt.bar(x, df['quality'], width=bar_width, label='Quality', color='lightgreen')
        plt.bar(x + bar_width, df['score'], width=bar_width, label='Final Score', color='coral')

        # Add threshold line
        plt.axhline(y=0.7, color='red', linestyle='--', label='Pass Threshold')

        # Customize plot
        plt.xlabel('ABET Criteria')
        plt.ylabel('Score')
        plt.title('ABET Compliance Scores by Criterion')
        plt.xticks(x, df['criterion'], rotation=45, ha='right')
        plt.ylim(0, 1.05)
        plt.legend()
        plt.tight_layout()

        # Save or display plot
        if output_path:
            plt.savefig(os.path.join(output_path, 'compliance_scores.png'))
            print(f"Visualization saved to {output_path}/compliance_scores.png")

        return plt

    def extract_evidence_by_criterion(self):
        """Extract evidence paragraphs grouped by criterion"""
        if self.classified_paragraphs is None:
            raise ValueError("No classified paragraphs available")

        # Filter out unknown and low confidence
        df = self.classified_paragraphs[
            (self.classified_paragraphs['criterion'] != 'Unknown') &
            (self.classified_paragraphs['confidence'] >= 0.2)
        ]

        evidence_by_criterion = {}
        for criterion, name in self.criteria_labels.items():
            criterion_df = df[df['criterion'] == criterion]

            if not criterion_df.empty:
                evidence_by_criterion[criterion] = {
                    'name': name,
                    'evidence': criterion_df[['paragraph', 'page', 'confidence']].to_dict('records')
                }
            else:
                evidence_by_criterion[criterion] = {
                    'name': name,
                    'evidence': []
                }

        return evidence_by_criterion

# Create output directory
!mkdir -p output

# Main execution
print("=== ABET Compliance Agent for Your PDF ===")

# Step 1: Upload the PDF file
from google.colab import files
import os

# Check if file already exists
if not os.path.exists('/content/ABET Self Study Report ME final.pdf'):
    print("Please upload your PDF file:")
    uploaded = files.upload()

    if 'ABET Self Study Report ME final.pdf' not in uploaded:
        print("File not found. Using the file name from uploaded files...")
        pdf_path = list(uploaded.keys())[0] if uploaded else None
    else:
        pdf_path = 'ABET Self Study Report ME final.pdf'
else:
    pdf_path = '/content/ABET Self Study Report ME final.pdf'
    print(f"Using existing file: {pdf_path}")

if not pdf_path:
    print("No PDF file was uploaded. Exiting.")
else:
    # Initialize the agent
    agent = ABETComplianceAgent()

    # Process the PDF document
    print("\n1. Extracting paragraphs from the PDF...")
    paragraphs = agent.extract_paragraphs_from_pdf(pdf_path)

    # Step 2: Classify the paragraphs (using rule-based method since we don't have a trained model)
    print("\n2. Classifying paragraphs to ABET criteria...")
    classified_paragraphs = agent.rule_based_classification(paragraphs)

    # Display summary of classification
    print("\nClassification Summary:")
    classification_summary = classified_paragraphs['criterion'].value_counts()
    print(classification_summary)

    # Step 3: Evaluate compliance
    print("\n3. Evaluating compliance with ABET criteria...")
    compliance_scores = agent.evaluate_compliance()

    # Step 4: Generate reports
    print("\n4. Generating reports...")
    report, non_compliance = agent.generate_compliance_report("./output")

    # Display overall result
    print(f"\nOverall compliance score: {report['overall']['score']:.2f}")
    print(f"Status: {report['overall']['status']}")

    # Display criterion scores
    print("\nCriterion scores:")
    for criterion, data in report['criteria'].items():
        print(f"{criterion} - {data['name']}: {data['score']:.2f} ({data['status']})")

    # Display non-compliance items
    if non_compliance:
        print("\nNon-compliance items (highest priority first):")
        for item in non_compliance:
            print(f"{item['criterion']} - {item['name']}: {item['gap_type']}, Need {item['evidence_needed']} more evidence items")
    else:
        print("\nNo non-compliance items found!")

    # Step 5: Visualize results
    print("\n5. Visualizing results...")
    agent.visualize_results("./output")

    # Step 6: Extract evidence by criterion for detailed review
    print("\n6. Extracting evidence by criterion...")
    evidence = agent.extract_evidence_by_criterion()

    # Display sample evidence for each criterion
    print("\nSample evidence for each criterion:")
    for criterion, data in evidence.items():
        print(f"\n{criterion} - {data['name']}:")
        evidence_items = data['evidence']
        if evidence_items:
            for i, item in enumerate(evidence_items[:2]):  # Show just 2 examples
                print(f"  - Evidence {i+1} (Page {item['page']}, Confidence: {item['confidence']:.2f}):")
                # Show truncated paragraph
                para = item['paragraph']
                if len(para) > 150:
                    para = para[:150] + "..."
                print(f"    {para}")
            if len(evidence_items) > 2:
                print(f"    ... and {len(evidence_items)-2} more evidence items")
        else:
            print("  No evidence found")

    print("\nAnalysis complete! All results are saved in the 'output' folder.")
    print("You can download the reports from the 'Files' tab in the left sidebar.")
```
